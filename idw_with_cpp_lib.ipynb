```python
import numpy as np
from idw import idw
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x_points = np.random.randint(0, 100, 20)
y_points = np.random.randint(0, 100, 20)
values = np.random.randint(1, 100, 20)

x_grid, y_grid = np.meshgrid(np.linspace(0, 100, 100), np.linspace(0, 100, 1005))
interpolated_values = np.zeros_like(x_grid)
```

```python
interpolated_values = idw(x_points, y_points, values, x_grid, y_grid, 2, 1e-6)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x_points, y_points, c=values, cmap='viridis', s=100, edgecolor='k', linewidth=1, alpha=0.8)
plt.contourf(x_grid, y_grid, interpolated_values, levels=50, cmap='viridis', alpha=0.8)
plt.colorbar(label='Interpolated Values')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Inverse Distance Weighted Interpolation with C++')
plt.show()
```
